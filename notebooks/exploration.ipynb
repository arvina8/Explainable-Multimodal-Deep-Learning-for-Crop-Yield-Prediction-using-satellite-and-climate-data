```python
# notebooks/exploration.ipynb

# --- 1. Setup ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load merged dataset
df = pd.read_csv("../data/processed/climate_yield_merged.csv")
df.head()
```

```python
print("Records:", len(df))
print("States:", df['state'].nunique())
print("Years:", df['year'].nunique())
print("Crops:", df['crop'].unique())

# Check for missing values
df.isnull().sum()
```

```python
# Example: Average temperature trend in Iowa
iowa = df[df['state'] == "IOWA"]
plt.figure(figsize=(10,5))
sns.lineplot(data=iowa, x="year", y="temp_mean_c", hue="crop")
plt.title("Iowa - Average Temperature Trend by Crop")
plt.show()
```

```python
# Yield distribution by crop
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="crop", y="yield_bu_acre")
plt.title("Yield Distribution by Crop")
plt.show()

# Yield trend over time
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x="year", y="yield_bu_acre", hue="crop")
plt.title("Yield Trends Over Time")
plt.show()
```

```python
# Scatter plot: Precipitation vs Yield
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="precipitation_mm", y="yield_bu_acre", hue="crop")
plt.title("Precipitation vs Yield")
plt.show()

# Correlation matrix
corr = df[["yield_bu_acre","temp_mean_c","precipitation_mm","humidity_pct","solar_radiation","wind_speed_ms"]].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Average climate features per crop
crop_climate = df.groupby("crop")[["temp_mean_c","precipitation_mm","humidity_pct"]].mean()
print(crop_climate)

# Boxplot of yield under different temperature ranges
df["temp_bin"] = pd.cut(df["temp_mean_c"], bins=[0,10,15,20,25,30])
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x="temp_bin", y="yield_bu_acre", hue="crop")
plt.title("Yield vs Temperature Range")
plt.show()
```
